```python


from huggingface_hub import hf_hub_download, login
from ultralytics import YOLO
import os, shutil, glob

login("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")

# Download model file from hugging face with token to login "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"
model_path = hf_hub_download(
    repo_id="rupesh002/Pothole_detcetion_yolo_model_120epochs_and_finetuned_50epochs",
    repo_type="model",
    filename="best (2).pt"
)

# Load  model
model = YOLO(model_path)
```

```python
def detect_potholes(video_file):
    if video_file is None:
        return None

    # Clean previous runs
    if os.path.exists("runs"):
        shutil.rmtree("runs")

    # Run YOLO
    model.predict(
        source=video_file,
        save=True,
        conf=0.35,
        iou=0.5
    )

    # Locate processed video
    output_videos = glob.glob("runs/detect/*/*.mp4")
    if len(output_videos) > 0:
        return output_videos[0]

    return None
```

```python
import gradio as gr

def detect_potholes_wrap(video):
    return detect_potholes(video)

with gr.Blocks() as demo:
    gr.Markdown("## üöß Pothole Detection App\nUpload a dashcam video to detect potholes using YOLO.")

    video_in = gr.Video(label="Upload Dashcam Video")
    result_video = gr.Video(label="Detected Output Video")

    run_btn = gr.Button("Detect Potholes")

    run_btn.click(fn=detect_potholes_wrap, inputs=video_in, outputs=result_video)

demo.queue().launch(share=True, debug=True)

```
